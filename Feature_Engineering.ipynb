### Now let's deal with the categorical data 

```python
catergorical_with_nan = [feature for feature in df.columns if df[feature].isnull().sum()>1 and df[feature].dtype == "O" ]
```

```python
catergorical_with_nan
```

```python
for feature in catergorical_with_nan:
    print('Missing Values {} is {} %'.format(feature, np.round(df[feature].isnull().mean(),4)))
```

```python
for feature in catergorical_with_nan:
    df[feature] = df[feature].fillna("Missing")
```

```python
for feature in catergorical_with_nan:
    print('Missing Values {} is {} %'.format(feature, np.round(df[feature].isnull().mean(),4)))
```

## Handling Rare Categorical Feature

We will remove categorical variables that are present less than 1% of the observations

```python
Catergorical_variables = [features for features in df.columns if df[features].dtype == 'O']
Catergorical_variables
```

```python
for feature in Catergorical_variables:
    temp = df.groupby(feature)['SalePrice'].count()/len(df)
    temp_df = temp[temp>0.01].index
    df[feature]= np.where(df[feature].isin(temp_df), df[feature], "Rare_Var")
```

```python
df[Catergorical_variables].head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
for feature in Catergorical_variables:
    df[feature] = le.fit_transform(df[feature])
```

```python
df.head()
```

## Now Let's moves towards the numerical variables

```python
numerical_variables = [features for features in df.columns if df[features].dtype != 'O' and df[features].isnull().sum()>0]
```

```python
numerical_variables
```

```python
for feature in numerical_variables:
    ## create a new feature to capture nan values
    df[feature+"_nan"] = np.where(df[feature].isnull(),1,0)
    df[feature].fillna(df[feature].median(), inplace=True)    
```

```python
df[numerical_variables].isnull().sum()
```

## Let's Capture the temporal variables
