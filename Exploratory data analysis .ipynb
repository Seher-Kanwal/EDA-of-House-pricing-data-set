```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('train.csv')
# as we need to show all columns 
pd.pandas.set_option('display.max_columns', None)
df.head()
```

# Here we will going to find the features that contains null values
### we first running a for loop if the features contains the null values we will calculating all the valuese and than in if condintion is that if any column contain missing value greater than display it as missing value

```python
features_with_na = [features for features in df.columns if df[features].isnull().sum()>1]
```

```python
features_with_na
```

## Values are between 0 and 1 like Poolqc contains 0.99% null values

```python
for features in features_with_na:
    print(features , 'contians',np.round(df[features].isnull().mean(),4),'% this much missing values ')
```

### Since they are many missing values, we need to find the relationship between missing values and Sales Price

Let's plot some diagram for this relationship

```python
for features in features_with_na:
    data = df.copy()
    # assign the value of 1 where the data is missing and 0 where data is present 
    data[features] = np.where(data[features].isnull(), 1, 0)  
    
    # lets calculate the means saleprice where the information is present or not
    
    # in features we have just two categories 0 and 1
    data.groupby(features)['SalePrice'].median().plot.bar()
    plt.title(features)
    plt.show()
    
```

#### In the first graph, we can see that the median of the sale price in greater in case of nan values,
#### we need to replace these values with some suitable data and so on for all columns

# Numerical Data

```python
features_with_numerical = [features for features in df.columns if df[features].dtypes != 'O']
print('the no of numerical data is:', len(features_with_numerical))
df[features_with_numerical].head()
```

## Temporal Variables(Eg: Datetime Variables)

**From the Dataset we have 4 year variables. We have extract information from the datetime variables like no of years or no of days. One example in this specific scenario can be difference in years between the year the house was built and the year the house was sold. We will be performing this analysis in the Feature Engineering which is the next video.**

```python
year_features = [feature for feature in features_with_numerical if 'Yr' in feature or 'Year' in feature]
year_features
```

# Content of these features

```python
for feature in year_features:
    print(feature)
    print(df[feature].unique())
```

## Lets analyze the Temporal Datetime Variables
## We will check whether there is a relation between year the house is sold and the sales price

```python
df.groupby("YrSold")["SalePrice"].median().plot()
plt.xlabel("year sold")
plt.ylabel('sale price')
plt.title('relationship between sale and year sold')
```

```python
for features in year_features:
    if features!= 'YrSold':
        data = df.copy()
        
        ## We will capture the difference between year variable and year the house was sold for
        ## for example if the year sold is 2008 and year built is 2003 than the differnce is 2008 - 2003 = 5
        ## we're actually creating if the age if the house is given than at how much it was sold
        data[features]=data['YrSold']-data[features]
        
        plt.scatter(data[features],data['SalePrice'])
        plt.xlabel(features)
        plt.ylabel('SalePrice')
        plt.show()
        
  
```
